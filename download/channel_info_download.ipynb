````python
# ==============================================
# å‰ç½®æ£€æŸ¥ä¸åˆå§‹åŒ–
# ==============================================
def init():
    # æ£€æŸ¥å…ƒæ•°æ®æ–‡ä»¶æ˜¯å¦å­˜åœ¨
    if not os.path.exists(METADATA_FILE):
        print(f"âŒ é”™è¯¯ï¼šå…ƒæ•°æ®æ–‡ä»¶ '{METADATA_FILE}' ä¸å­˜åœ¨ï¼")
        sys.exit(1)
    
    # æ£€æŸ¥æ–‡ä»¶æ˜¯å¦ä¸ºç©º
    if os.path.getsize(METADATA_FILE) == 0:
        print(f"âŒ é”™è¯¯ï¼šå…ƒæ•°æ®æ–‡ä»¶ '{METADATA_FILE}' ä¸ºç©ºï¼")
        sys.exit(1)
    
    # åˆ›å»ºä¸‹è½½ç›®å½•ï¼ˆä¸å­˜åœ¨åˆ™åˆ›å»ºï¼‰
    if not os.path.exists(DOWNLOAD_PATH):
        os.makedirs(DOWNLOAD_PATH)
        print(f"âœ… åˆ›å»ºä¸‹è½½ç›®å½•ï¼š{DOWNLOAD_PATH}")
    
    # åˆå§‹åŒ– Synapse å®¢æˆ·ç«¯å¹¶ç™»å½•
    print("\nğŸ” æ­£åœ¨åˆå§‹åŒ– Synapse å®¢æˆ·ç«¯...")
    syn = synapseclient.Synapse() 
    syn.login(authToken=Auth_TOKEN) 
    try:
        # å°è¯•æ— äº¤äº’ç™»å½•ï¼ˆå·²ç™»å½•è¿‡åˆ™ç›´æ¥ç”Ÿæ•ˆï¼‰
        syn.login(silent=True)
        print("âœ… Synapse ç™»å½•æˆåŠŸï¼")
    except SynapseError:
        # æ— ç¼“å­˜ç™»å½•ä¿¡æ¯æ—¶ï¼Œæ‰‹åŠ¨è¾“å…¥è´¦å·å¯†ç 
        print("â„¹ï¸  æœªæ£€æµ‹åˆ°ç™»å½•ç¼“å­˜ï¼Œè¯·è¾“å…¥ Synapse è´¦å·å¯†ç ï¼š")
        username = input("ç”¨æˆ·åï¼š")
        password = input("å¯†ç ï¼š")
        try:
            syn.login(username=username, password=password, rememberMe=True)
            print("âœ… Synapse ç™»å½•æˆåŠŸï¼")
        except Exception as e:
            print(f"âŒ ç™»å½•å¤±è´¥ï¼š{str(e)}")
            sys.exit(1)
    
    return syn

# ==============================================
# è¯»å–å¹¶ç­›é€‰ TSV æ•°æ®
# ==============================================
def load_and_filter_metadata():
    print(f"\nğŸ“Š æ­£åœ¨è¯»å–å…ƒæ•°æ®æ–‡ä»¶ï¼š{METADATA_FILE}")
    try:
        # è¯»å– TSV æ–‡ä»¶ï¼ˆsep='\t' åˆ¶è¡¨ç¬¦åˆ†éš”ï¼Œheader=0 ç¬¬ä¸€è¡Œä¸ºè¡¨å¤´ï¼‰
        df = pd.read_csv(
            METADATA_FILE,
            sep='\t',
            header=0 if SKIP_HEADER else None,
            encoding='utf-8',
            dtype=str  # é¿å… Synapse ID è¢«è§£æä¸ºæ•°å­—
        )
        
        # æ£€æŸ¥å¿…è¦åˆ—æ˜¯å¦å­˜åœ¨
        required_cols = [DATA_ACCESS_COL_NAME, SYNAPSE_ID_COL_NAME]
        missing_cols = [col for col in required_cols if col not in df.columns]
        if missing_cols:
            print(f"âŒ é”™è¯¯ï¼šæœªæ‰¾åˆ°å¿…è¦åˆ—ï¼š{', '.join(missing_cols)}")
            print(f"è¯·æ£€æŸ¥è¡¨å¤´åç§°æ˜¯å¦æ­£ç¡®ï¼ˆå½“å‰è¡¨å¤´ï¼š{', '.join(df.columns)}ï¼‰")
            sys.exit(1)
        
        # ç­›é€‰æ¡ä»¶ï¼šData Access ä¸ä¸ºç©ºä¸”ä¸ç­‰äº 'Coming Soon'ï¼ˆå¿½ç•¥ç©ºæ ¼ï¼‰
        df['Data Access_trimmed'] = df[DATA_ACCESS_COL_NAME].str.strip()
        filtered_df = df[
            (df['Data Access_trimmed'].notna()) & 
            (df['Data Access_trimmed'] != 'Coming Soon')
        ].copy()
        
        # å»é™¤ Synapse ID ä¸ºç©ºçš„è¡Œ
        filtered_df = filtered_df[filtered_df[SYNAPSE_ID_COL_NAME].str.strip().notna()]
        
        # å»é‡ï¼ˆé¿å…é‡å¤ä¸‹è½½ï¼‰
        filtered_df = filtered_df.drop_duplicates(subset=SYNAPSE_ID_COL_NAME, keep='first')
        
        print(f"âœ… æ•°æ®ç­›é€‰å®Œæˆï¼š")
        print(f"   - åŸå§‹æ•°æ®è¡Œæ•°ï¼š{len(df)}")
        print(f"   - ç¬¦åˆæ¡ä»¶è¡Œæ•°ï¼š{len(filtered_df)}")
        
        if len(filtered_df) == 0:
            print("â„¹ï¸  æ²¡æœ‰æ‰¾åˆ°ç¬¦åˆæ¡ä»¶çš„ Synapse IDï¼Œè„šæœ¬é€€å‡º")
            sys.exit(0)
        
        return filtered_df[SYNAPSE_ID_COL_NAME].tolist()  # è¿”å›æ‰€æœ‰ç¬¦åˆæ¡ä»¶çš„ Synapse ID
    
    except Exception as e:
        print(f"âŒ è¯»å–/ç­›é€‰æ•°æ®å¤±è´¥ï¼š{str(e)}")
        sys.exit(1)

# ==============================================
# ä¸‹è½½ Synapse èµ„æº
# ==============================================
def download_synapse_resources(syn, synapse_ids):
    print(f"\nğŸš€ å¼€å§‹ä¸‹è½½ {len(synapse_ids)} ä¸ªèµ„æºï¼ˆä¿å­˜åˆ°ï¼š{DOWNLOAD_PATH}ï¼‰")
    success_count = 0
    fail_count = 0
    fail_ids = []
    
    for idx, syn_id in enumerate(synapse_ids, 1):
        syn_id = syn_id.strip()
        print(f"\n[{idx}/{len(synapse_ids)}] æ­£åœ¨ä¸‹è½½ï¼š{syn_id}")
        
        try:
            # ä¸‹è½½èµ„æºåˆ°æŒ‡å®šç›®å½•
            entity = syn.get(
                syn_id,
                downloadLocation=DOWNLOAD_PATH,
                ifcollision='skip'# 'overwrite'  # é‡å¤æ–‡ä»¶è¦†ç›–ï¼ˆå¯æ”¹ä¸º 'skip' è·³è¿‡ï¼‰
            )
            print(f"âœ… ä¸‹è½½æˆåŠŸï¼š{entity.name}ï¼ˆä¿å­˜è·¯å¾„ï¼š{entity.path}ï¼‰")
            success_count += 1
        except Exception as e:
            print(f"âŒ ä¸‹è½½å¤±è´¥ï¼š{str(e)}")
            fail_count += 1
            fail_ids.append(syn_id)
    
    # è¾“å‡ºä¸‹è½½ç»Ÿè®¡
    print(f"\nğŸ“Š ä¸‹è½½å®Œæˆï¼")
    print(f"âœ… æˆåŠŸï¼š{success_count} ä¸ª")
    print(f"âŒ å¤±è´¥ï¼š{fail_count} ä¸ª")
    if fail_ids:
        print(f"âŒ å¤±è´¥çš„ Synapse IDï¼š{', '.join(fail_ids)}")

# ==============================================
# ä¸»å‡½æ•°
# ==============================================
def main():
    print("=" * 50)
    print("ğŸ“¥ Synapse èµ„æºæ‰¹é‡ä¸‹è½½è„šæœ¬")
    print("=" * 50)
    
    # åˆå§‹åŒ–
    syn = init()
    
    # è¯»å–å¹¶ç­›é€‰æ•°æ®
    synapse_ids = load_and_filter_metadata()
    
    # æ‰¹é‡ä¸‹è½½
    download_synapse_resources(syn, synapse_ids)
    
    print("\n" + "=" * 50)
    print("ğŸ‰ è„šæœ¬æ‰§è¡Œå®Œæˆï¼")
    print("=" * 50)

if __name__ == "__main__":
    main()
````
